```python
# 1. Working with IP Addresses

# Exercise 1: Extend the script to calculate:
# - Broadcast address
# - First and last usable host addresses
# - Number of usable hosts
# - Change the IP address to have it CIDR prefix ( e.g. /24)
# and compare resulting networks

# import ipaddress
# def analyse_ip(ip_str):
# # Create an IP interface object
# ip = ipaddress.ip_interface(ip_str)
# print(f"Address: {ip.ip}")
# print(f"Network: {ip.network}")
# print(f"Netmask: {ip.netmask}")
# print(f"Is private: {ip.ip.is_private}")
# print(f"Is global: {ip.ip.is_global}")
# # List all hosts in the network
# if ip.network.num_addresses < 256: # Only for small networks
# print("\nHosts in network:")
# for host in ip.network.hosts():
# print(host)
# # Example usage
# analyse_ip('192.168.1.1')

import ipaddress
def analyse_ip(ip_str):
# Create an IP interface object
    ip = ipaddress.ip_interface(ip_str)

    print(f"Address: {ip.ip}")
    print(f"Network: {ip.network}")
    print(f"Netmask: {ip.netmask}")
    print(f"Is private: {ip.ip.is_private}")
    print(f"Is global: {ip.ip.is_global}")

    # broadcast address
    broadcast = ip.network.network_address 
    print(f"Broadcast address: {broadcast}")

    # First and last usable host addresses
    first_host = ip.network.network_address + 1
    last_host = ip.network.broadcast_address - 1
    print(f"First usable host: {first_host}")
    print(f"Last usable host: {last_host}")

    # Number of usable hosts
    number_of_hosts = ip.network.num_addresses - 2
    print(f"Number of usable hosts: {number_of_hosts}")

    # Change the IP address to have it CIDR prefix ( e.g. /24)
    new_cidr = 24
    new_ip = ipaddress.ip_interface(f"{ip.ip}/{new_cidr}")
    print(f"\nChanged IP with CIDR /{new_cidr}: {new_ip}")
    print(f"New Network: {new_ip.network}")

    # Compare resulting networks
    if ip.network == new_ip.network:
        print("The networks are the same")
    else:
        print("The networks are different")
    
    # End of modified section

    # List all hosts in the network
    if ip.network.num_addresses < 256: # Only for small networks
        print("\nHosts in network:")
    for host in ip.network.hosts():
        print(host)

# Example usage
analyse_ip('192.168.1.1/24')   
```

```python
# Exercise 4: (Challenge) Create a subnetting plan for a company with 4
# departments requiring the following hosts:
# • Engineering: 30 hosts
# • Marketing: 15 hosts
# • Finance: 10 hosts
# • HR: 5 hosts

# Use the network address 172.16.0.0/16 and determine appropriate subnet masks for
# each department

import ipaddress

departments = {
    "Engineering": 30,
    "Marketing": 15,
    "Finance": 10,
    "HR": 5
}

# Function to calculate the required subnet size
def subnet_size(hosts):
    # Add 2 for network and broadcast addresses
    required_size = hosts + 2
    
    prefix_length = 32 - (required_size - 1).bit_length()
    return prefix_length


# Start with the base network
network = ipaddress.ip_network('172.16.0.0/16')

# Initialize the subnet iterator
current_subnet = network.subnets(new_prefix=24)

# Allocate subnets for each department
for department, hosts in departments.items():
    # Calculate the required subnet size
    prefix_length = subnet_size(hosts)
    
    # Find the next subnet that can accommodate the required size
    while True:
        subnet = next(current_subnet)
        if subnet.prefixlen <= prefix_length:
            break
    
    # Adjust the subnet to the required prefix length
    subnet = list(subnet.subnets(new_prefix=prefix_length))[0]
    
    # Print 
    print(f"Department: {department}")
    print(f"Subnet: {subnet}")
    print(f"Network: {subnet.network_address}")
    print(f"Netmask: {subnet.netmask}")
    print(f"Broadcast: {subnet.broadcast_address}")
    print(f"Hosts: {subnet.num_addresses - 2}")
    print()


```

```python
# b. Simulate DHCP operations
# dhcp_simple.py - Simplified DHCP Simulator
# Server Configuration
server = {
    "ip_pool": ["192.168.1.100", "192.168.1.101", "192.168.1.102"],
    "leases": {}
}

# Client Configuration
client = {
    "mac": "AA:BB:CC:DD:EE:FF",
    "ip": None
}

def send_discover():
    print("\n[CLIENT] Step 1: Sending DHCP DISCOVER")
    return {
        "type": "DISCOVER",
        "mac": client["mac"]
    }

def make_offer(discover):
    print("\n[SERVER] Step 2: Making DHCP OFFER")
    if not server["ip_pool"]:
        print("No IPs available!")
        return None
    offered_ip = server["ip_pool"].pop(0)
    return {
        "type": "OFFER",
        "mac": discover["mac"],
        "ip": offered_ip
    }

def send_request(offer):
    print("\n[CLIENT] Step 3: Sending DHCP REQUEST")
    return {
        "type": "REQUEST",
        "mac": offer["mac"],
        "ip": offer["ip"]
    }

def send_ack(request):
    print("\n[SERVER] Step 4: Sending DHCP ACK")
    server["leases"][request["mac"]] = request["ip"]
    return {
        "type": "ACK",
        "mac": request["mac"],
        "ip": request["ip"]
    }

def main():
    print("=== Simple DHCP Simulation ===")
    # Client starts process
    discover = send_discover()
    # Server responds
    offer = make_offer(discover)
    if not offer:
        return
    # Client continues
    request = send_request(offer)
    # Server finalizes
    ack = send_ack(request)
    # Update client IP
    client["ip"] = ack["ip"]
    print("\n=== Result ===")
    print(f"Client {client['mac']} got IP: {client['ip']}")
    print("Server leases:", server["leases"])

if __name__ == "__main__":
    main()
```
