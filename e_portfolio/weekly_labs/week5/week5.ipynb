```python
# Initial code given 

# 1. Building a Simple TCP Server

import socket

# Create a TCP socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(('localhost', 65432))
server_socket.listen(1) # Allow 1 pending connection
print("TCP Server is listening...")

while True:
    client_socket, client_address = server_socket.accept()
    print(f"Connected to {client_address}")

    data = client_socket.recv(1024)
    print(f"Received: {data.decode()}")

    # Echo back the data
    client_socket.sendall(b"ACK: " + data)
    client_socket.close()


```

```python
# Exercise 6: (Challenge) Implement chat application for 2 clients

# server
import socket
import threading

clients = []

def handle_client(client_socket):
    while True:
        try:
            message = client_socket.recv(1024)
            if not message:
                break
            print(f"Received: {message.decode()}")
            broadcast(message, client_socket)
        except:
            clients.remove(client_socket)
            break

def broadcast(message, sender_socket):
    for client in clients:
        if client != sender_socket:
            client.sendall(message)

server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(('localhost', 65432))
server_socket.listen(5)

print("Chat server started...")
while True:
    client_socket, addr = server_socket.accept()
    print(f"New connection from {addr}")
    clients.append(client_socket)
    threading.Thread(target=handle_client, args=(client_socket,)).start()

client_socket.close()
```

```python
# Exercise 7: Add encryption for secure messaging (use the cryptography library)

# server
import socket
import threading
from cryptography.fernet import Fernet

key = Fernet.generate_key()
cipher = Fernet(key)

clients = []

def handle_client(client_socket):
    while True:
        try:
            encrypted_message = client_socket.recv(1024)
            if not encrypted_message:
                break
            message = cipher.decrypt(encrypted_message).decode()
            print(f"Decrypted: {message}")
            broadcast(encrypted_message, client_socket)
        except:
            clients.remove(client_socket)
            break

def broadcast(message, sender_socket):
    for client in clients:
        if client != sender_socket:
            client.sendall(message)

server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
server_socket.bind(('localhost', 65432))
server_socket.listen(5)

print("Secure chat server started...")
while True:
    client_socket, addr = server_socket.accept()
    print(f"New connection from {addr}")
    clients.append(client_socket)
    client_socket.sendall(key)  # Send encryption key to client
    threading.Thread(target=handle_client, args=(client_socket,)).start()
    
client_socket.close()
server_socket.close()
```

```python
# client
import socket
import threading
from cryptography.fernet import Fernet

client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client_socket.connect(('localhost', 65432))

key = client_socket.recv(1024)
cipher = Fernet(key)

def receive_messages():
    while True:
        try:
            encrypted_message = client_socket.recv(1024)
            print("\nDecrypted:", cipher.decrypt(encrypted_message).decode())
        except:
            break

threading.Thread(target=receive_messages).start()

while True:
    message = input("Enter message: ")
    encrypted_message = cipher.encrypt(message.encode())
    client_socket.sendall(encrypted_message)

client_socket.close()
```

```python
# Exercise 8: Fetch weather data from an API and send it over TCP.
# server
import socket

server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
server_socket.bind(('localhost', 65432))
server_socket.listen(1)

print("Weather data server is running...")
client_socket, _ = server_socket.accept()

weather_data = client_socket.recv(1024).decode()
print(f"Received weather data: {weather_data}")

client_socket.close()
server_socket.close()

```

```python
# client
import socket
import requests

api_url = "https://api.open-meteo.com/v1/forecast?latitude=51.47&longitude=-0.0363&current_weather=true"
response = requests.get(api_url)
weather_data = response.json()['current_weather']

client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client_socket.connect(('localhost', 65432))

client_socket.sendall(str(weather_data).encode())

print("Weather data sent!")
client_socket.close()
```
